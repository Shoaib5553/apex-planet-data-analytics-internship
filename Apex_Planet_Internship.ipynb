```python
import pandas as pd
import sqlite3
```

```python
# Loading the csv files into dataframe
orders_dataset = pd.read_csv('/content/cleaned_orders_dataset.csv')
order_items_dataset = pd.read_csv('/content/cleaned_order_items_dataset.csv')
customers_dataset = pd.read_csv('/content/cleaned_customers_dataset.csv')
products_dataset = pd.read_csv('/content/cleaned_products_dataset.csv')
geolocation_dataset = pd.read_csv('/content/cleaned_geolocation_dataset.csv')
sellers_dataset = pd.read_csv('/content/cleaned_sellers_dataset.csv')
order_payments = pd.read_csv('/content/cleaned_order_payments.csv')
order_reviews = pd.read_csv('/content/cleaned_order_reviews_dataset.csv')
product_category_name_translation = pd.read_csv('/content/cleaned_product_category_name_translation.csv')
```

```python
# Creating the database file
conn = sqlite3.connect('olist_ecommerce.db')
```

```python
# Converting all the dataframes into the database
orders_dataset.to_sql('orders', conn, if_exists='replace', index=False)
order_items_dataset.to_sql('order_items', conn, if_exists='replace', index=False)
customers_dataset.to_sql('customers', conn, if_exists='replace', index=False)
products_dataset.to_sql('products', conn, if_exists='replace', index=False)
geolocation_dataset.to_sql('geolocation', conn, if_exists='replace', index=False)
sellers_dataset.to_sql('sellers', conn, if_exists='replace', index=False)
order_payments.to_sql('order_payments', conn, if_exists='replace', index=False)
order_reviews.to_sql('order_reviews', conn, if_exists='replace', index=False)
product_category_name_translation.to_sql('product_category_name_translation', conn, if_exists='replace', index=False)
```

```python
# Checking our SQL Queries
query = """
SELECT * FROM orders
LIMIT 5;
"""

# Run the query and display the output as a neat table
pd.read_sql_query(query, conn)
```

```python
# Query 1.	What are the top 5 product categories by total revenue generated?

query = '''
SELECT product_category_name_english, SUM(price) AS "Total Revenue"
FROM products, order_items, product_category_name_translation
WHERE order_items.product_id = products.product_id AND products.product_category_name = product_category_name_translation.product_category_name
GROUP BY product_category_name_english
LIMIT 5;
'''

pd.read_sql_query(query, conn)
```

```python
# Query 2. What is the monthly order volume trend? (Are sales growing or shrinking month-over-month)

query = '''
SELECT strftime('%Y-%m', order_purchase_timestamp) AS "Purchase Month", COUNT(order_id) AS "Order Volume"
FROM orders
GROUP BY "Purchase Month"
ORDER BY "Purchase Month";
'''

pd.read_sql_query(query, conn)
```

```python
# Query 3.	Which 5 customer cities generate the highest total revenue for the platform?

query = '''

SELECT customer_city, SUM(price) AS "Total Revenue"
FROM customers, order_items, orders
WHERE customers.customer_id = orders.customer_id AND orders.order_id = order_items.order_id
GROUP BY customer_city
ORDER BY customer_city
LIMIT 5;
'''

pd.read_sql_query(query, conn)
```

```python
# Query 4.	What is the average order value (AOV) for each distinct payment method?

query = '''

SELECT payment_type, AVG(price) AS "Average Order Value"
FROM order_payments, order_items
WHERE order_payments.order_id = order_items.order_id
GROUP BY payment_type;
'''

pd.read_sql_query(query, conn)
```

```python
# Query 5.	What is the monthly customer acquisition trend? (How many unique customers made their first purchase each month?)

query = '''

SELECT
    strftime('%Y-%m', first_purchase_date) AS "Month",
    COUNT(customer_unique_id) AS "New Customers Acquired"

FROM (
    SELECT customer_unique_id, MIN(order_purchase_timestamp) AS first_purchase_date
    FROM orders, customers
    WHERE orders.customer_id = customers.customer_id
    GROUP BY customer_unique_id
) AS FirstPurchases

GROUP BY Month
ORDER BY Month;
'''

pd.read_sql_query(query, conn)
```

```python
# Query 6.	How many orders were delivered late (past the estimated delivery date), and what percentage of total delivered orders does this represent?

query = '''

SELECT COUNT(*) AS "Total Late Deliveries", COUNT(*) * 100.0 / (SELECT COUNT(*)
                                                                FROM orders
                                                                WHERE order_status = 'delivered') AS "Percentage of Total Late Deliveries"
FROM orders
WHERE order_delivered_customer_date > order_estimated_delivery_date;
'''

pd.read_sql_query(query, conn)
```

```python
# Query 7.	What are the top 5 specific products (by product_id) that have the highest total number of 1-star reviews?

query = '''

SELECT product_id, COUNT(*) AS "Total 1-Star Reviews"
FROM order_reviews, order_items
WHERE order_reviews.order_id = order_items.order_id AND review_score = 1
GROUP BY product_id
ORDER BY "Total 1-Star Reviews" DESC
LIMIT 5;
'''

pd.read_sql_query(query, conn)
```
